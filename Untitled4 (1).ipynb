```python
import pandas as pd

# Load data from Excel file
def load_data_from_file(filename="/content/fashion_department_store_data.xlsx"):
    df = pd.read_excel(filename)
    items = list(df.itertuples(index=False, name=None))
    return items

# Insertion Sort (Iterative)
def insertion_sort(items):
    n = len(items)
    for i in range(1, n):
        key = items[i]
        j = i - 1
        while j >= 0 and items[j][1] > key[1]:  # Sort by price
            items[j + 1] = items[j]
            j -= 1
        items[j + 1] = key

# Merge Sort (Recursive)
def merge_sort(items):
    if len(items) > 1:
        mid = len(items) // 2
        left_half = items[:mid]
        right_half = items[mid:]

        merge_sort(left_half)
        merge_sort(right_half)

        i = j = k = 0

        while i < len(left_half) and j < len(right_half):
            if left_half[i][1] < right_half[j][1]:  # Sort by price
                items[k] = left_half[i]
                i += 1
            else:
                items[k] = right_half[j]
                j += 1
            k += 1

        while i < len(left_half):
            items[k] = left_half[i]
            i += 1
            k += 1

        while j < len(right_half):
            items[k] = right_half[j]
            j += 1
            k += 1

# Iterative Search
def iterative_search(items, target_name):
    for item in items:
        if item[0] == target_name:
            return item
    return None

# Recursive Search
def recursive_search(items, target_name, index=0):
    if index >= len(items):
        return None
    if items[index][0] == target_name:
        return items[index]
    return recursive_search(items, target_name, index + 1)

if __name__ == "__main__":
    data = load_data_from_file()

    while True:
        print("\n1. Sort Items by Price (Ascending) - Iterative")
        print("2. Sort Items by Price (Ascending) - Recursive")
        print("3. Search for an Item (Iterative)")
        print("4. Search for an Item (Recursive)")
        print("5. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            insertion_sort(data)
            print("Top 10 Items sorted by price (ascending) - Iterative:")
            for item in data[:10]:
                print(item)

        elif choice == "2":
            merge_sort(data)
            print("Top 10 Items sorted by price (ascending) - Recursive:")
            for item in data[:10]:
                print(item)

        elif choice == "3":
            target = input("Enter the item name to search (Iterative): ")
            result = iterative_search(data, target)
            print(f"Result: {result}" if result else "Item not found.")

        elif choice == "4":
            target = input("Enter the item name to search (Recursive): ")
            result = recursive_search(data, target)
            print(f"Result: {result}" if result else "Item not found.")

        elif choice == "5":
            print("Exiting the program.")
            break

        else:
            print("Invalid choice. Please try again.")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Fungsi untuk kompleksitas waktu
def insertion_sort_complexity(n):
    return n ** 2

def merge_sort_complexity(n):
    return n * np.log2(n)

def iterative_search_complexity(n):
    return n

def recursive_search_complexity(n):
    return n

# Data untuk elemen (ukuran input)
n = np.linspace(1, 300, 300)  # Range untuk 300 elemen

# 1. Grafik Sorting
insertion_sort_y = insertion_sort_complexity(n)
merge_sort_y = merge_sort_complexity(n)

plt.figure(figsize=(10, 6))
plt.plot(n, insertion_sort_y, label="Insertion Sort (O(n^2))", color="red")
plt.plot(n, merge_sort_y, label="Merge Sort (O(n log n))", color="orange")

plt.fill_between(n, 0, insertion_sort_y, color="red", alpha=0.2)
plt.fill_between(n, 0, merge_sort_y, color="orange", alpha=0.2)

plt.title("Sorting Algorithms Complexity (n = 300)", fontsize=16)
plt.xlabel("Number of Elements (n)", fontsize=12)
plt.ylabel("Operations", fontsize=12)
plt.legend(fontsize=12)
plt.grid(True)
plt.show()

# 2. Grafik Searching
iterative_search_y = iterative_search_complexity(n)
recursive_search_y = recursive_search_complexity(n)

plt.figure(figsize=(10, 6))
plt.plot(n, iterative_search_y, label="Iterative Search (O(n))", color="blue")
plt.plot(n, recursive_search_y, label="Recursive Search (O(n))", color="green")

plt.fill_between(n, 0, iterative_search_y, color="blue", alpha=0.2)
plt.fill_between(n, 0, recursive_search_y, color="green", alpha=0.2)

plt.title("Searching Algorithms Complexity (n = 300)", fontsize=16)
plt.xlabel("Number of Elements (n)", fontsize=12)
plt.ylabel("Operations", fontsize=12)
plt.legend(fontsize=12)
plt.grid(True)
plt.show()
```
